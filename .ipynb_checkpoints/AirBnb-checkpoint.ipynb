reviews_per_month: The average number of reviews per month. This column is a numeric variable.<br>

```python
# Numerical Variables.
num_columns = data[['price', 'minimum_nights','number_of_reviews','reviews_per_month','availability_365','calculated_host_listings_count']]
```

```python
# Creating a function that gives the mean, mode, median, std, and variance
def descriptive_stats(num_columns):
    num_var = []
    for i in num_columns:
        if data[i].dtype in ['int64', 'float64']:  # Check if column is numerical
            mean = data[i].mean()
            median = data[i].median()
            mode = data[i].mode().values[0]
            variance = data[i].var()
            standard_deviation = data[i].std()
            num_var.append([i, mean,median, mode, variance, standard_deviation])
    nums = pd.DataFrame(num_var, columns=['Feature', 'Mean','Median', 'Mode', 'Variance', 'Standard Deviation'])
    return nums
```

```python
# Calling that function.
descriptive_stats(num_columns)
```

**Investigate the relationship between room type, neighborhood group, and price. Perform appropriate statistical tests (e.g., t-test, ANOVA) to determine if there are significant differences in rental prices based on room type and neighborhood group.**

H0 - There is no significant difference in rental prices between room type and neighbourhood group<br>
H1 - There is a significant difference in rental prices between room type and neighbourhood group

```python
# Importing the stats Library
import scipy.stats as stats
```

```python
# t-test for room type
room_type_groups = data.groupby('room_type')
entire_home_apt_prices = data[data['room_type'] == 'Entire home/apt']['price']
private_room_prices = data[data['room_type'] == 'Private room']['price']
ttest, pval = stats.ttest_ind(entire_home_apt_prices, private_room_prices)
if pval < 0.05:
    print('There is a significant difference in rental prices between entire home/apt and private room listings')
else:
    print('There is no significant difference in rental prices between entire home/apt and private room listings')
```

```python
# t-test for neighborhood group
neighbourhood_groups = data.groupby('neighbourhood_group')
manhattan_prices = data[data['neighbourhood_group'] == 'Manhattan']['price']
brooklyn_prices = data[data['neighbourhood_group'] == 'Brooklyn']['price']

ttest, pval = stats.ttest_ind(manhattan_prices, brooklyn_prices)
if pval < 0.05:
    print('There is a significant difference in rental prices between Manhattan and Brooklyn listings')
else:
    print('There is no significant difference in rental prices between Manhattan and Brooklyn listings')
```

```python
# ANOVA for room type
shared_room_prices = data[data['room_type'] == 'Shared room']['price']
anova = stats.f_oneway(entire_home_apt_prices, private_room_prices, shared_room_prices)
if pval < 0.05:
    print('There is a significant difference in rental prices based on room type')
else:
    print('There is no significant difference in rental prices based on room type')
```

```python
# ANOVA for neighborhood group
bronx_prices = data[data['neighbourhood_group'] == 'Bronx']['price']
brooklyn_prices = data[data['neighbourhood_group'] == 'Brooklyn']['price']
manhattan_prices = data[data['neighbourhood_group'] == 'Manhattan']['price']
queens_prices = data[data['neighbourhood_group'] == 'Queens']['price']
staten_island_prices = data[data['neighbourhood_group'] == 'Staten Island']['price']

anova = stats.f_oneway(manhattan_prices, brooklyn_prices, queens_prices, bronx_prices, staten_island_prices)
if pval < 0.05:
    print('There is a significant difference in rental prices based on neighborhood group')
else:
    print('There is no significant difference in rental prices based on neighborhood group')
```

We can see that all the tests that we performed shows that there is a significance difference in rental prices based on room type and neighborhood group
